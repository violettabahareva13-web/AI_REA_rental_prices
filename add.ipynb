```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('team_A_data.csv')
pd.set_option('display.max_columns', None)
df.head()
```

```python
df["ceiling_bin"] = pd.cut(df["ceiling_height_m"], bins=10)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    x='ceiling_bin',
    y='price_rub',
    errorbar=None
)
plt.title('Зависимость цены аренды от высоты потолков')
plt.xlabel('Высота потолков, м')
plt.ylabel('Цена аренды, руб')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df_filtered = df[df['passenger_elevator'] <= 6]
sns.barplot(data=df_filtered, x='passenger_elevator', y='price_rub', errorbar=None)

plt.title('Зависимость цены аренды от наличия пассажирского лифта')
plt.xlabel('Наличие пассажирского лифта, кол-во')
plt.ylabel('Цена аренды, руб')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='cargo_elevator', y='price_rub', errorbar=None)
plt.title('Зависимость цены аренды от наличия грузового лифта')
plt.xlabel('Наличие грузового лифта, кол-во')
plt.ylabel('Цена аренды, руб')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
df["area_bin"] = pd.cut(df["area_m2"], bins=12)

plt.figure(figsize=(14, 6))
sns.barplot(
    data=df,
    x='area_bin',
    y='price_rub',
    errorbar=None
)

plt.title('Зависимость цены аренды от площади квартиры')
plt.xlabel('Площадь квартиры, м2')
plt.ylabel('Цена аренды, руб')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
